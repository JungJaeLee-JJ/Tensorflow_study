### How to minimize cost?
Hypothesis H(x) = Wx 일때
1. cost(W) $$\frac{1}{m} \sum(W*x_i-y_i)^2$$
      1-1. W=1, cost(w)? 0 

      1-2. w=0, cost(w)? $$\frac{1}{3}(1+4+9)$$

**x가 W고 Y가 Cost로 하여 Cost가 최저인 점을 찾으면된다.**

### Gradient descent algorithm
1. start at any value
2. 접선(기울기)에서 cost가 감소하는 방향으로 이동한다.

    2-1. cost 함수를 아래와 같다고 할때(2는 미분을 편하게 하기위해 나눠준다)
    $$\frac{1}{2m} \sum(W*x_i-y_i)^2$$
    
    2-2. 이를 미분하였을때(단 여기서 알파는 leaning late)
    $$W=W - \alpha\frac{1}{m}\sum(W*x_i-y_i)x_i$$

    2-3. 문제점? -> **만약 시작점이 다를때, 도착하는 최저의 W값이 달라질수있다**
    
    
**따라서 Cost function이 Convex function(오목한 형태라서 어디서 시작해서 결국 같은 W값으로 산정됨)일때 Gradient descent algorithm을 사용 가능하다**




```python
#import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() #버젼 문제 때문에, placeholder가 없다고 나와 이와 같이 해결함
import matplotlib.pyplot as plt

X= [1,2,3]
Y= [1,2,3]

W=tf.placeholder(tf.float32)

hypothesis =X*W

cost = tf.reduce_mean(tf.square(hypothesis-Y))

sess = tf.Session()
sess.run(tf.global_variables_initializer())

W_val=[]
cost_val=[]
for i in range(-30,50):
    feed_W = i*0.1
    curr_cost,curr_W=sess.run([cost,W],feed_dict={W:feed_W})
    W_val.append(curr_W)
    cost_val.append(curr_cost)
    
plt.plot(W_val,cost_val)
plt.show()
    
 
```

```python
x_data = [1,2,3]
y_data = [1,2,3]

W = tf.Variable(tf.random_normal([1]),name='weight')
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

hypothesis = X * W

cost = tf.reduce_sum(tf.square(hypothesis -Y))

#Gradient Descent
learning_rate=0.1
gradient = tf.reduce_mean((W*X-Y)*X)
descent = W-learning_rate*gradient
update = W.assign(descent)
  
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(21):
    sess.run(update,feed_dict={X: x_data,Y: y_data})
    print(step,sess.run(cost,feed_dict={X: x_data,Y: y_data}), sess.run(W))
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

X= [1,2,3]
Y= [1,2,3]

w=tf.Variable(5.0)
hypothesis = X*w
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(100):
    print(step,sess.run(w))
    sess.run(train)
```

### 변수가 복수일때? ###

$$H(x_1,x_2,x_3) = w_1x_1 + w_2x_2 + w_3x_3 + b$$

**변수가 많이 질수록 식이 길어진다 -> 행렬로 해결하자**

**------>H(X) = XW**


```python
x1_data = [73.,93.,89.,96.,73.]
x2_data = [80.,88.,91.,98.,66.]
x3_data = [75.,93.,90.,100.,70]
y_data = [152.,185.,180.,196.,142.]

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

w1 =tf.Variable(tf.random_normal([1]),name='weight1')
w2 =tf.Variable(tf.random_normal([1]),name='weight2')
w3 =tf.Variable(tf.random_normal([1]),name='weight3')
b =tf.Variable(tf.random_normal([1]),name='bias')

hypothesis = x1*w1 + x2*w2 + x3*w3 +b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(2001):
    cost_val,hy_val,_=sess.run([cost,hypothesis,train],
                              feed_dict={x1: x1_data,x2: x2_data,x3: x3_data, Y:y_data})
    if step%10 == 0:
        print(step, "Cost : ",cost_val,"\nPrediction\n",hy_val)
```

```python
tf.set_random_seed(777)

x_data = [[73.,80.,75.],[93.,88.,93.],[89.,91.,90.],[96.,98.,100.],[73.,66.,70]]
y_data = [[152.],[185.],[180.],[196.],[142.]]

print(x_data,y_data)

X= tf.placeholder(tf.float32,shape = [None,3])
Y= tf.placeholder(tf.float32,shape = [None,1])

W=tf.Variable(tf.random_normal([3,1]),name='weight')
b=tf.Variable(tf.random_normal([1]),name='bias')

hypothesis=tf.matmul(X,W)+b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(2001):
    cost_val,hy_val,_=sess.run([cost,hypothesis,train],feed_dict={X: x_data, Y: y_data})
    if step%10 == 0:
        print(step, "Cost : ",cost_val,"\nPrediction\n",hy_val)
```
